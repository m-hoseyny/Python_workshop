```python
class News:
    def __init__(self):
        self.text = None
        self.title = None
        self.date = None
        self.link = None
        self.full_text = None
    def __repr__(self):
        return "{},\n{}\n{}\n{}".format(self.title, self.text, self.date, self.link)
    
    def check_keywords(self, keywords = ['دلار', 'آمریکا', 'تحریم']):
        for item in keywords:
            if item in self.text or item in self.title:
                self.get_complete_news()
        
    def get_complete_news(self):
        url = self.link
        page = urllib2.urlopen(url)
        soup = BeautifulSoup(page, 'lxml')
        text_raw = soup.body.main.div.div.div.article('div')[2].div('div')[1].find_all('p')
        full_text = ''
        for text in text_raw:
            full_text += text.text.strip() + '\n'
        self.full_text = full_text
    
class NewsIsna(News):
    pass
```

```python
news_list = []
for news in list_news_raw:
    local_news = News()
    local_news.text = news.p.text.strip()
    local_news.title = news.h3.text.strip()
    local_news.date = news.time.text.strip()
    local_news.link = 'https://www.isna.ir' + news.h3.a['href'].strip()
    news_list.append(local_news)
```

```python
class NewsDataBase:
    def __init__(self, path = 'news.db'):
        self.conn = sqlite3.connect(path)
        cursor = self.conn.cursor()
        try:
            cursor.execute("""CREATE TABLE news (title VARCHAR(1000), summary VARCHAR(1000), full_text VARCHAR(1000), date VARCHAR(30), keywords TEXT, 
                veiws INT)""")
        except:
            pass
        cursor.close()
    def insert_news(self, news):
        cursor = self.conn.cursor()
        cursor.execute("""INSERT INTO news (title, summary, date) VALUES('{}', '{}', '{}')""".format(\
                                                                            news.title, news.text, news.date))
        cursor.close()
    def insert_news_list(self, list_news):
        for news in list_news:
            self.insert_news(news)
    def get_all_news(self):
        cursor = self.conn.cursor()
        cursor.execute("""SELECT * FROM news""")
        news = cursor.fetchall()
        return news
        cursor.close()
```

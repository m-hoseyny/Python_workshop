```python
news_raw = soup.body.main.div.div.div.div.contents
print(news_raw)
```

```python
list_news_raw = news_raw[5].find_all('li')
print(list_news_raw)
print(len(list_news_raw))
```

```python
print('اولین تیتر خبرها')
print((list_news_raw[2]).h3.text.strip())

print('متن خبر')
print((list_news_raw[2]).p.text.strip())

print('link')
print((list_news_raw[2]).h3.a['href'])

print('datetime')
print((list_news_raw[2]).time.text.strip())

site_url = 'https://www.isna.ir'

class News:
    title = None
    text = None
    news_date = None
    link = None
    
news_list = []

for news in list_news_raw:
    local_news = News()
    local_news.title = news.h3.text.strip()
    local_news.text = news.p.text.strip()
    local_news.time = news.time.text.strip()
    local_news.link = site_url + news.h3.a['href']
    news_list.append(local_news)
```

```python
for news in news_list:
    news_str = news.title + '\n' + news.text + '\n' + news.time + '\n' + news.link
    print(news_str)
```
